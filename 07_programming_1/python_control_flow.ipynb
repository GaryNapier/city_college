# `if`, `else`, `elif`

We can take all out operators that result in Boolean data types, i.e. something that results in `True` or `False`

These are:

`and`
`or`
`not`
`==` - is equal to
`!=` - not equal to
`is`
`is not`
`>` - greater than
`<` - less than
`>=` - greater than or equal to
`<=` - less than or equal to

We can say in Python:
> "if this condition is `True`, then do this"


```python
if 10 > 1:
    print("Hello")
```

A couple of things to note about this `if` statement:

- The *condition* `10 > 1` evaluates to `True`, so the code inside the `if` statement executes
- The code inside the `if` statement needs to be **indented** with four spaces, otherwise it's not proper Python syntax and you'll get an error
    - You can also use tab instead of four spaces but tabs can cause problems and the preferred standard is four spaces - https://www.reddit.com/r/learnpython/comments/8cann8/tabs_vs_spaces_i_dont_get_it/
    - In reality most people use tabs because it's just one key... 
    - In VSCode, it seems to put four spaces in automatically when you press `return` (this might be down to one of the add-ins like Pylint)

What if the condition was `False` and wanted to do something else? 

### `else`

We can extend the condition with `else`:

> "if this condition is `True`, then do this... If it's not `True` (i.e. `False`), then do something else."

```python
if 1 > 10:
    print("Hello")
else:
    print("Goodbye")
```

### `elif`

You can extend this even further with multiple conditions with `elif`. 

These are multiple mutually exclusive conditions (i.e. only one condition can be true). 

It stops checking ("breaks") as soon as it finds a `True` condition.

```python
score = 75

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
```

### `elif... else`

You can combine `elif` and `else` - if it doesn't find any `True` conditions then finally do something else.. 

```python
score = 59

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
else: 
  print("Fail")
```

### `if/else` and types

Be careful that the type of object you're supplying to the condition statement(s) can be meaningfully evaluated. 

In the code below, the `score` variable is being passed as a string. 

It doesn't mean anything to ask if a string is "greater than or equal to 90", so you get an error. 

```python
score = "59"

if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
else: 
  print("Fail")
```

### Complexity and refactoring

As with our Boolean section, you can make the conditons as complex as you like, as long as they evaluate to `True` or `False`:

```python
if (
    (
        not(10 > 100) or 
        (True or True)
    )
      and (3 >= 4)
    ):
    print("Hello")
else:
    print("Goodbye")
```

Of course, this isn't very easy to read, so it's better to evaluate before passing to the `if` statement. 

If your code starts to look ugly and it's hard to read, then it's best to stop and think about how to make it nearer and clearer. 

This is called *refactoring* - when the code does the same thing, but is written in a different, improved way. 

```python
x = not(10 > 100)
y = (True or True)
z = 3 >= 4

x_or_y = x or y
x_or_y_and_z = x_or_y and z

if x_or_y_and_z:
    print("Hello")
else:
    print("Goodbye")
```

# `match` / `case`

Alternative to `if`/`else`. Sometimes called `case/when` or `case/switch` in other languages.

Instead of writing many if..else statements, you can use the match statement.

```python
status = 418

match status:
    case 400:
        print("Bad request")
    case 404:
        print("Not found")
    case 418:
        print("Server down")
    # If an exact match is not confirmed, this last case will be used if provided ("default" case)
    case _:
        print("Something's wrong with the internet")
```

### Combine Values

Use the pipe character `|` as an `or` operator in the `case` evaluation to check for more than one match:

```python
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5: # "1 or 2 or 3 or 4 or 5"
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

### If Statements as Guards

You can add `if` statements in the `case` evaluation as an extra condition-check. 

```python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")
  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
  case _:
    print("No match")
```

# `for` loop

`for` loops *iterate* (go through one-by one) over one of our collection types - list, tuple, dictionary, set.

You can also loop over strings, which is one our basic types. 

Anything that can be looped over is called an *iterable*. 

`for` loops execute statements, once for each item in a collection (or string). 

Here, the `x` variable stands for each one of the members in the `fruits` list, in turn:

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

Note - like the `if`, the code inside the loop has to be indented

### Loop over strings

```python
for x in "banana":
    print(x)
```

### Combine with `if`

We can put an `if` in the loop to check if any of the members meet a condition. 

Only print the member if it's "banana"

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        print(x)
```

### `break` statement

Stop the loop before it has looped over all items. 

Exit the loop when x is "banana":

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

Exit the loop when x is "banana", but this time the break comes before the print:

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
```

### `continue` statement

Stop the *current* iteration, and continue with the next:

Do not print banana:

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
```

### `range()` function

Loop through a set of code a specified number of times, use the `range()` function,

`range()` returns a sequence of numbers, **starting from 0** by default, and increments by 1 (by default), and ends at a specified number.

```python
for x in range(6):
    print(x)
```

### Nested Loops

A loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop":

Print each adjective for every fruit:

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y) 
```

You can nest as many loops as you like, but it's rare to get three nests, and any more than this suggests there's probably a better way to do it. 

### `pass` statement

`for` loops cannot be empty, but if you for some reason have a `for` loop with no content, put in `pass` to avoid an error.

```python
for x in [0, 1, 2]:
    pass 
```

# `while` loop

Execute a set of statements *as long as* a condition is true.

Print i as long as i is less than 6:

```python
i = 1
while i < 6:
    print(i)
    i += 1 # add 1 to i and re-assign the new value to i
```

### `while` and `break`

Stop the loop even if the while condition is true.

Exit the loop when i is 3:

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

### `continue` statement

Stop the current iteration, and continue with the next.

Skip printout of 3:

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

# Functions

Functions are the "unit" of programming. It's a shortcut to packaging up lots of code 

- Start making them early
- Start making them often
- Make them small and build up more complex functions from smaller ones
    - The smaller they are the easier to test and fix
    - This fits with our "modular" approach to programming/CS
- Rules of thumb:
    - If you find yourself repeating a chunk of code more than a couple of times, make a function
    - If your code that just does one thing is becoming long and complicated, make a function

Here's some code before making a function. If it starts to look like this, then make a function:

```python
temp1 = 77
celsius1 = (temp1 - 32) * 5 / 9
print(celsius1)

temp2 = 95
celsius2 = (temp2 - 32) * 5 / 9
print(celsius2)

temp3 = 50
celsius3 = (temp3 - 32) * 5 / 9
print(celsius3) 
```

Apart from being repetive and tedious to go through, it's more likely to have mistakes. 

The numbers 32, 5 and 9 are **hardcoded** here (they're not assigned to variables). 

Any repitition of hardcoded variables is more likely to be wrong (even if copied and pasted). 

Also, if you have to change one of the parameters (e.g. 32), you'd have to *go through every hardcoded value* and change it. 

So we either assign to variables, or, better, put in a function. 

Defined using the `def` keyword, followed by a function name and parentheses (`()`), then the code (indented).

*Run* the function by calling the name with the paretheses:

```python
def my_function():
    print("Hello from a function")

my_function()
```

This is not an interesting function because it just outputs something, but takes no inputs and doesn't do anything with those inputs. 

Add inputs as *arguments* insode the paretheses `()`.

Then do something with those arguments and output the result.

The line with `return` will output whaterver comes after it. 

Here `fahrenheit` is the *argument* to the function, i.e. the input.

Inside the function `fahrenheit` is a variable that gets used in the equation to get celsius. 

```python
def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50)) 
```

Use `=` in the arguments to assign a default value. 

This way the argument doesn't have to be explicitly passed in (like the 77 value below). 

Putting any other value in (77 here) overwrites the default value. 

```python
def fahrenheit_to_celsius(fahrenheit = 212):
    return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius())
print(fahrenheit_to_celsius(77))
```
