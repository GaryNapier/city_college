### The "shebang" line

What went wrong?

This time `getoutput()` is running `hello.py` *without* the `python` part.

Why do we test this?

We want to make the program "executable", which means it runs as a stand-alone program without calling another program before it. So in the terminal, we just want to run:

`./hello.py`

In the `hello.py` file, we need to add a "shebang" line (on the first line) starting `#!`. 

Then we follow it with `/usr/bin/env python`, so the whole line is:

`#!/usr/bin/env python`

Notice this is the first line we had in the `test.py` file.. 

Explanation:

- Tell the operating system to use the `env` program (which is found in the directory `root` (`/`), `usr`, `bin`) to find the `python` program.

### $PATH

If you run `which python` in the terminal, you'll see the location where Python is being run from. 

This is a path, and we have a special variable called `$PATH` where `env` or `which` are looking. 

Run `echo $PATH` in the terminal - should get something like:

```
/usr/local/bin:/usr/bin:/bin:/usr/sbin:/sbin:/Library/TeX/texbin:/opt/X11/bin:/Applications/quarto/bin:/...etc
```

This is a list of directories separated by `:`. 

To find a program, the system goes through these directories in turn, until it finds it and runs it. 

You can mimic this directly in the terminal with `env python` - run this in terminal and Python should start (type `quit()` to exit). 

So the whole "shebang" line says "use `env` to start the `python` program and then execute the rest of the code in the file".

### Executability

If you remember from the Unix/command line lesson, we can run `ls -alh` and it will list file permissions in the left column:

```
drwxr-xr-x  6 <username>  staff   192B Jun 15 14:55 .pytest_cache
drwxr-xr-x  3 <username>  staff    96B Jun 15 15:31 __pycache__
-rw-r--r--  1 <username>  staff    12K Jun 15 16:44 hello.ipynb
-rw-r--r--  1 <username>  staff    24B Jun 15 13:54 hello.py
```

The fourth place (after the `rw`, standing for read/write) has either an `x` or a `-`. And `x` means it's executable and so can run stand-alone.

If you look at the error we got from the test, it was a *permissions* error. So we have to give permissions to make the file executable (along with the shebang line). 

Run `chmod +x hello.py` to make the file executable. 

- `chmod` means "change mode"
- Note that we're using `+` instead of `-` for the flag

Now you should see an `x` in the file listing:

`ls -alh`
`-rwxr-xr-x  1 <username>  staff    24B Jun 15 13:54 hello.py`

Now run `./hello.py` in the terminal and it should just run (don't forget the shebang line!).

### $PATH again

`hello.py` is executable, but *only if you're in the same directory*.

If you move directory and try `./hello.py` it says `bash: ./hello.py: No such file or directory`. 

Or if you try to run it like a regular program like `ls` you get `command not found`.

You *can* run if you specify the path:

```
cd ../ # move directory one up
01_hello/hello.py
```

But this can be a pain to keep specifying the directory. With more complex programs the files could be spread out in a complex structure. 

So what we can do is copy the file to somewhere that is is *already* on the path. Then the system will find the file and execute it. 

For example, `/usr/local/bin` is one of the directories in `$PATH` (it's common for programs to be stored there).

So we could do `cp hello.py /usr/local/bin` and then we can just run `hello.py` from anywhere. 

Also running `which hello.py` will find the program, because `usr/local/bin` is on the `$PATH`:

```
which hello.py
/usr/local/bin/hello.py
```
...

But We might not want to do this because we might not know what these core system files are really for, so we can make our own dirctory e.g. `~/bin` or `~/programs`. 

`mkdir ~/programs`

Now we can add this directory to the `$PATH`:

`PATH=~/programs:$PATH`

This command is saying "reassign the `PATH` variable to `~/programs` and join together with whatever `PATH` already is (separated by `:`). 

Now run `echo $PATH` again, and your new directory should be at the start:

```
~/programs:/usr/local/bin:/usr/bin:/bin:/usr/sbin:/sbin:/Library/TeX/texbin:/opt/X11/bin:/Applications/quarto/bin:/...etc
```

Now the program can be run from anywhere by just calling `hello.py`.

`which` will also find the program:

```
which hello.py
~/programs/hello.py
```

Ok, that's all you need to know about the `$PATH` for now. If you want to add your programs to the PATH then do so, but we'll just continue to run from the terminal in the same dir as the program.
