# Part 2 - running tests

## Step 1

Make a test file `test.py` in the same directory as the `hello.py` file.

## Step 2

Opening lines of code in `test.py`:

````python
#!/usr/bin/env python3

"""Tests for hello.py"""

import os
from subprocess import getstatusoutput, getoutput

prg = './hello.py'
````

I'll explain the first line and the `import` lines later, but for now we just need to know that we're telling out test file that the file we're testing is called `hello.py` (and the `./` part tells `test.py` that it's in **this** directory, i.e. the same directory we're running `test.py` from).

## Step 3

The first test we'll run tests if the program exists, so we write a function to do that:

````python
def test_exists():
    """exists"""

    assert os.path.isfile(prg)
````

We're going to be running a program in the terminal called `pytest`. 

It will use the `test.py` file and look for functions that begin with `test_`, so here our function starts with that. 

The `assert` keyword will return `True` or `False`. It's like asking "Is this statement `True`?"

`os` is using the `os` package. This is a way of getting Python to interact with the terminal and your system in general (https://docs.python.org/3/library/os.html).

`os.path` is using the `path` method from the `os` package to look in the current path (i.e. where `test.py` is saved). 

`os.path.isfile` is a method asking if it's `True` that a file exists in the current path. 

So `assert` together with `os.path.isfile` is asking "Is it True that the the file exists?".

So our first test of `hello.py` is actually just testing if the file exists in the first place.

Run `pytest test.py`

## Step 4

"Test runnable" - does the program run using Python 3?

````python
def test_runnable():
    """Runs using python3"""

    out = getoutput(f'python3 {prg}')
    assert out.strip() == 'Hello, World!'
````

The `getoutput()` function runs the orignal code we ran in the terminal to run `hello.py`, i.e. `python3 hello.py`. 

It then checks that the output is the string `"Hello, World!"`.

Add this function to `test.py` and run again. 

## Step 5

"Test executable"

````python
def test_executable():
    """Says 'Hello, World!' by default"""

    out = getoutput(prg)
    assert out.strip() == 'Hello, World!'
````

### Executability and the "shebang" line

What went wrong?

This time `getoutput()` is running `hello.py` *without* the `python3` part.

Why do we test this?

We want to make the program "executable", which means it runs as a stand-alone program without calling another program before it. So in the terminal, we just want to run:

`./hello.py`
