## Step 3 - Program exists

The first test we'll run tests if the program exists, so we write a function to do that:

````python
def test_exists():
    """exists"""

    assert os.path.isfile(prg)
````

We're going to be running a program in the terminal called `pytest`.

`os` is using the `os` package. This is a way of getting Python to interact with the terminal and your system in general (https://docs.python.org/3/library/os.html).

`os.path` is using the `path` method from the `os` package to look in the current path (i.e. where `test.py` is saved). 

`os.path.isfile` is a method asking if it's `True` that a file exists in the current path. 

So `assert` together with `os.path.isfile` is asking "Is it True that the the file exists?".

So our first test of `hello.py` is actually just testing if the file exists in the first place.

# Part 3 - Adding a parameter and help

The program just outputs one thing "Hello, World!", which isn't really a program. It takes no inputs. 

We need to add *arguments* to the program, i.e. inputs. 

## Step 1 - add `import argparse`

Add `import argparse` to `hello.py` (before the `print()` line).

`argparse` is a library that allows you to add arguments to the program from the command line. 

(https://docs.python.org/3/library/argparse.html)

So now we can run the program, but pass something else and it will do something with it, like:

```
hello.py Universe
Hello, Universe!
```
or
```
hello.py Mum
Hello, Mum!
```

## Step 2 - add argparse code

Let's add the argparse code and go through each line:

````python
#!/usr/bin/env python
# Say hello

import argparse

parser = argparse.ArgumentParser(description='Say hello')
parser.add_argument('name', help='Name to greet')
args = parser.parse_args()
print('Hello, ' + args.name + '!')
````

`parser = argparse.ArgumentParser(description='Say hello')`

- Create an `ArgumentParser` *object* assigned to `parser` variable.
- We'll use this to handle all the arguments. 
- We'll learn more about objects later, but for now just need to include this code without thinking too much about it. 
- Also sets the help message (`description`). 

`parser.add_argument('name', help='Name to greet')`

- Use the `add_argument` method on the `parser` object.
- Adds an argument to the program. 
- We give the argument a name (in this case the arg name is "name", confusingly, but it could be anything as we'll see in future problems).
- Also add the help message. 

`args = parser.parse_args()`

- Use the `parse_args()` method to extract all the arguments.
- Our "name" argument is now an *attribute* of `args` (`args.name`).
- In this case, it will be a *string*, whatever the user has passed to the command line as the argument, e.g. `hello.py Mum`, "Mum" is the string now stored in `args.name`

`print('Hello, ' + args.name + '!')`

- Now we actually do something with the argument. 
- Here we *concatenate* the string inside `args.name` with the other two strings "Hello, " and "!", using the *operator* `+`. 
- Finally print the concatenated string with `print()` function!

# Part 4 - Making the arg optional

What if we want to run the program with default values (no args)?

## Step 1 - change `add_argument()` method 

```
parser.add_argument('-n', '--name', metavar='name',
                    default='World', help='Name to greet')
```

- Add the short flag `-n` and long flag `--name`
- Now we need to add `metavar=` - will show up in the usage to describe the arg. 
- `default=` allows no args to be suppied, and the "name" arg will automatically be the string "World!". 

## Step 2 - run without/with args

Runs like it did before:

```

```


```
$ ./hello.py -h

  -n name, --name name  Name to greet
```

### Arguemnts - notes

- Args with dashes are optional 
    - can be left out
    - *may* have default values
    
- Parameters without dashes are *positional*
    - usually required, so no default values

# Part 5 - more tests

## Step 1 - test usage

```
def test_usage():
    """usage"""

    for flag in ['-h', '--help']:
        status, out = getstatusoutput(f'{prg} {flag}')
        assert status == 0
        assert out.lower().startswith('usage')
```

Let's go through the lines... 

### `status, out = getstatusoutput(f'{prg} {flag}')`

#### `getstatusoutput()` 

    - running our program again, for each of the flags `-h` and `--help`
    - has two outputs in a *tuple*:

`(0, 'usage: hello.py [-h] [-n name]\n\nSay hello\n\noptional arguments:\n  -h, --help  show this help message and exit\n  -n name, --name name  Name to greet')`

    - The first output, `0` is the status of the command line output ("standard out" or "stdout"). In the background, every time a command is run, there's an "exit status". 
    - Exit `0` typically signifies success and any non-zero value indicates an error. 
    - So here it's output `0`, so that means output was successful. 
    - [Exit status handling](https://tiunnts.medium.com/mastering-output-and-exit-status-handling-in-bash-scripts-7e99ec1a4635)

    - The second output (starting `usage`) is the actual output of the program. 

#### `assert status == 0`

    - This line now checks that the first output is indeed `0` (indicating no errors), returning `True` or `False`.

#### `assert out.lower().startswith('usage')`

    - Next we check the second output (`out`). 
    - Here we're not checking the whole message, just that it starts with "usage". But then we know the help message has run correctly with `argparse`. 

## Step 2 - test input

Next we want to test our new argument (input) `name`. 

```
def test_input():
    """test for input"""

    for val in ['Universe', 'Multiverse']:
        for option in ['-n', '--name']:
            status, out = getstatusoutput(f'{prg} {option} {val}')
            assert status == 0
            assert out.strip() == f'Hello, {val}!'
```

#### `for val in ['Universe', 'Multiverse']:`

    - Check a couple of input strings, these could be anything

#### `for option in ['-n', '--name']:`

    - Test both of the flags

#### `status, out = getstatusoutput(f'{prg} {option} {val}')`

    - Use the `getstatusoutput()` function again to get status and output
    - Except now we pass it a diffent command line line - the program name `hello.py`, each of the flags, and our two test strings, in a *nested for-loop*.

#### `assert status == 0`

    - Same as our `test_usage()` function. 

#### `assert out.strip() == f'Hello, {val}!'`

    - Now we want to check if the output is "Hello, <something>", whatever inputs we have passed it, in this case "Universe" and "Multiverse".
