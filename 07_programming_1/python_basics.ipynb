## Boolean (True or false)

Sometimes called "logical" values. 

There are operators in Python to test equality:

`and`
`or`
`not`
`==` - is equal to
`!=` - not equal to
`is`
`is not`
`>` - greater than
`<` - less than
`>=` - greater than or equal to
`<=` - less than or equal to

Each statement with a logical operator will evaluate to True or False:

```python
print(True and False)
print(True or False)
print(not True)
print(10 > 100)
print(10 < 100)
print(10 >= 10)
print(10 <= 10)
print(10 == 100)
print(100 == 100)
print(10 != 100)
```

You can have quite complex logical operations embedded in parentheses which will each be evaluated in turn (though they're hard to follow so you would probably want to evaluate each chunk and assign to a variable before passing the the operators):

```python
print((True and False) or (True or False))
print((True or False) and (True or False))
print(
    (
        not(
            (10 > 100) or 
            (True or True)
        )
    ) and
        (3 >= 4)
    )
```

You can also evaluate strings for equality. 

```python
print("Hello" == "Hello")
```

You can also: 

- mix up evaluation of strings and numbers 

```python
print("Hello" == 7)
```

- evaluate strings as being greater than or less than another string

```python
print("Apple" < "Banana")
```

- but you can't ask if a string is less/greater than a number

```python
print("Hello" > 7)
```

Be careful with the strange behaviour in the cases of:
`"Hello" == 7`
and
`"Apple" < "Banana"`
where you *don't* get an error. 

These evaluations don't really mean anything (except maybe in certain contexts), and because you don't get an error you might not have wanted to compare a string to a number for example. 

Aside:
Why can we compare a string to a number, or compare less/greater than on two strings?

`==` and `!=` are defined for *any* two objects.
`<`, `>`, `<=`, `>=` only work when the objects have an ordering relationship defined.

Strings actually do have an ordering relationship because behind the scenes Python is converting the strings to numbers. 

```python
print("a" < "b")
print("a" > "b")
```

Anyway, the main point here is that all of these evaluations evaluate to `True` or `False`, which makes then **Boolean** types of object.  

```python
print(type(True))
print(type(False))
print(type(True and False))
print(type(True or False))
print(type(10 > 100))
print(type(10 < 100))
print(type(10 == 100))
print(type(100 == 100))
print(type(100 != 100))
```

Difference Between `is` and `==`

- `is` - Checks if both variables point to the same object in memory
- `==` - Checks if the values of both variables are equal

```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
```

# Putting objects together ("collection data types")

We can't do very much if we deal with one object at a time, so we need ways of collecting them togther, which will in turn create another type of object.

Here are the data collection types:

- **List** - ordered and changeable. Allows duplicate members.
- **Tuple** - ordered and unchangeable. Allows duplicate members.
- **Set** - unordered, unchangeable (except add/remove), and unindexed. No duplicate members.
- **Dictionary** - is a collection which is ordered and changeable. No duplicate members.

## Lists

Ordered and changeable. Allows duplicate members.

A list is just a collection of any type of item, including mixed types. 

### Lists - square brackets

You can create a list with square brackets `[]`:

```python
x = ["Hello", "Goodbye"]
y = ["Hello", 3, True]
print(x)
print(y)
print(type(x), type(y))
```

You can also have embedded lists or any other type of object as one of the members of the list:

Here, `x` is a list the last member is a list. 
In `y`, the last member is a **set**, which is one of our collections types. 

But overall, both objects are just lists as you can see from the `type()` printout.

```python
x = ["Hello", 3, True, ["x", "y", "z"]]
y = ["Hello", 3, True, {"x", "y", "z"}]
print(type(x), type(y))
```

### Lists - `list()` function

Or use the `list()` function. 

To use this you must actually pass the `list()` function a collection of objects, i.e. a tuple, set or dictionary (which are described below).

```python
x = list(("Hello", 3)) # List from tuple
y = list({"Hello", 3}) # List from set
z = list(
    {
        "House num": 123, 
        "Street": "Ipswich Rd"
    }
) # List from dictionary
print(x)
print(y)
print(z)
print(type(x), type(y), type(z))
```

### Lists - access

Pull out members of lists with the list **indices** in square brackets `[]` immediately after the list. 

```python
x = ["Hello", 3, True, ["x", "y", "z"]]
y = ["Hello", 3, True, {"x", "y", "z"}]
print(x[1])
print(y[3])
```

Here, even though we used **1** and **3**, we actually pulled out the **second** and **fourth** members of the list. Why?

For reasons I've never understood, Python starts with 0 when counting things. It's completely unhelpful and counter-intuitive but we're stuck with it.

## Tuples

Ordered and unchangeable. Allows duplicate members.

Think of tuples as "one thing", unlike lists which are collections of things. 

That is, tuples are useful when you want to see if something matches something else, or is in soemthing else, in the exact way it appears in the tuple. 

### Tuples - round brackets `()`

Create tuples with round brackets `()`:

```python
x = ("Hello", "Goodbye")
y = ("xyz", 3, ["Apple", "Banana"])
print(x)
print(y)
print(type(x), type(y))
```

### Tuples - `tuple()` function

Like lists, tuple has its own function. 

Again, you have to pass what is already a *collection* of items to the function:

```python
a = tuple(("Apples", "Bananas", "Pears")) # Tuple from tuple
b = tuple(["Apples", "Bananas", "Pears"]) # Tuple from list
c = tuple({"Apples", "Bananas", "Pears"}) # Tuple from set
d = tuple(    
        {
            "House num": 123, 
            "Street": "Ipswich Rd"
        }
    ) # Tuple from dictionary
print(a)
print(b)
print(c)
print(d)
print(type(a), type(b), type(c), type(d))
```

### Tuples - access

Access members just like lists:

```python
x = ("Apples", "Bananas", "Pears")

x[0]
```

## Sets

Unordered, unchangeable (except add/remove), and unindexed. No duplicate members.

These are used mainly when you want to do logical operations on collections of items - e.g. finding overlaps between collections. 

Is one member of a set inside another set?
Are all of these members of a set not in this other set?

### Sets - curly braces `{}`

```python
x = {"Apples", "Bananas", True, 1, 10}

print(x)
print(type(x))
```

Why has `1` disappeared from the set when it's printed?

This is because we can't have **duplicate** members. 

But how is `1` a duplicate member? There's only one `1`. 

This is because Python (like most languages) treats  `1` and `0` as `True` and `False`, respectively in some contexts. 

Similarly, if we make a set with `False` and `0`, then only one will get used (the first to appear).

```python
{"Apples", "Bananas", True, 1, 10, 0, False}
```

### Sets - `set()` function

```python
a = set(("Apples", "Bananas", "Pears")) # set from tuple
b = set(["Apples", "Bananas", "Pears"]) # set from list
c = set({"Apples", "Bananas", "Pears"}) # set from set
d = set(    
        {
            "House num": 123, 
            "Street": "Ipswich Rd"
        }
    ) # set from dictionary
print(a)
print(b)
print(c)
print(d)
print(type(a), type(b), type(c), type(d))
```

### Sets - access

You can't access sets in the same way as the other collection types because they are **unordered**. 

```python
x = {"Apples", "Bananas", True, 1, 10, 0, False}

x[2]
```

## Dictionaries

Used to store data values in `key:value` pairs.

Ordered, changeable, no duplicates.

### Dictionaries - curly braces `{}`

Like sets, we can use the curly braces, but the content is different, this time we have `key:value`. 

```python
x = {"House num": 123, "Street": "Ipswich Rd", "City": "Norwich"}
print(x)
print(type(x))
```

You can have any data type as the `value`, and it's common to see lists here:

```python
x = {"House num": 123, "Street": "Ipswich Rd", "City": "Norwich", "pets": ["rabbit", "fish", "dog"]}
print(x)
```

### Dictionaries - `dict()` function

There is a function for dictionaries but you can't really use it like the other functions because you need `key:value` pairs. 

The first line here errors because it doesn't know what to do with the set to convert it to a dictionary. 

```python
dict(("apples", "pears", "bananas"))
dict({"apples", "pears", "bananas"})
dict(["apples", "pears", "bananas"])
```

But you can create one like this:

```python
dict(House_num = 123, Street = "Ipswich Rd", City = "Norwich", Pets = ["rabbit", "fish", "dog"])
```

Notice that the antecedents of the `=` are not strings, they are more like **variables**. 

Hence, we have to use `House_num` with the underscore to keep it together. 

### Dictionaries - access

Access in dictionaries works a bit differently because now we have the `key:value` pairs. 

Subset the dictionary with names of keys:

```python
x = {"House num": 123, "Street": "Ipswich Rd", "City": "Norwich", "pets": ["rabbit", "fish", "dog"]}
x["pets"]
```

## `len()` function

One useful function is `len()` which returns the length of any data type collection

```python
print(len(["apples", "bananas", "pears"]))
print(len(("apples", "bananas", "pears")))
print(len({"apples", "bananas", "pears", False, 0}))
print(
    len(
        {
            "Shop": "Sainsbury's", 
            "Fruit": ["apples", "bananas", "pears"]
        }
    )
)
```
