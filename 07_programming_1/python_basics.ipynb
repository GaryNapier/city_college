# Values, objects and data types

"Values", "objects" and "data types" are essentially just different names for any "thing" in Python. 

But every "thing" belongs to a type, or class. 

See - https://www.w3schools.com/python/python_datatypes.asp

Use the `type()` function to find out what type an object belongs to:

```python
type(3)
```

```python
type("Hello")
```

The basic types of thing we can have in Python are numbers (which come in a few different types), strings (text, basically), and Booleans (True or False).

## Numbers

### Integers

```python
print(1)
print(10)
print(13)
print(-10000)
```

```python
print(type(1))
print(type(10))
print(type(13))
print(type(-10000))
```

### Floating point numbers ('float')

```python
print(type(0.11))
print(type(0.0001))
print(type(4.212))
print(type(3.1415))
```

### Complex numbers (don't worry about these)

```python
print(type(1j))
```

## Strings 

Anything wrapped in double (") or single (') quotes.

```python
print(type("Hello world"))
print(type("xyx"))
print(type('Hello world'))
```

## Boolean (True or false)

Sometimes called "logical" values. 

There are operators in Python to test equality:

`and`
`or`
`==` - is equal to
`!=` - not equal to
`>` - greater than
`<` - less than
`>=` - greater than or equal to
`<=` - less than or equal to

Each statement with a logical operator will evaluate to True or False:

```python
print(True and False)
print(True or False)
print(10 > 100)
print(10 < 100)
print(10 >= 10)
print(10 <= 10)
print(10 == 100)
print(100 == 100)
print(10 != 100)
```

You can have quite complex logical operations embedded in parentheses which will each be evaluated in turn (though they're hard to follow so you would probably want to evaluate each chunk and assign to a variable before passing the the operators):

```python
print((True and False) or (True or False))
print((True or False) and (True or False))
print(
    (
        (10 > 100) or 
        (True or True)
    ) and
        (3 >= 4)
    )
```

You can also evaluate strings for equality. 

```python
print("Hello" == "Hello")
```

You can also: 

- mix up evaluation of strings and numbers 

```python
print("Hello" == 7)
```

- evaluate strings as being greater than or less than another string

```python
print("Apple" < "Banana")
```

- but you can't ask if a string is less/greater than a number

```python
print("Hello" > 7)
```

Be careful with the strange behaviour in the cases of:
`"Hello" == 7`
and
`"Apple" < "Banana"`
where you *don't* get an error. 

These evaluations don't really mean anything (except maybe in certain contexts), and because you don't get an error you might not have wanted to compare a string to a number for example. 

Aside:
Why can we compare a string to a number, or compare less/greater than on two strings?

`==` and `!=` are defined for *any* two objects.
`<`, `>`, `<=`, `>=` only work when the objects have an ordering relationship defined.

Strings actually do have an ordering relationship because behind the scenes Python is converting the strings to numbers. 

```python
print("a" < "b")
print("a" > "b")
```

Anyway, the main point here is that all of these evaluations evaluate to `True` or `False`, which makes then **Boolean** types of object.  

```python
print(type(True))
print(type(False))
print(type(True and False))
print(type(True or False))
print(type(10 > 100))
print(type(10 < 100))
print(type(10 == 100))
print(type(100 == 100))
print(type(100 != 100))
```

## `NA` and `NaN` (or `nan`)

`NA` = "Not Applicable/Available"
`NaN`/`nan` = "Not a Number"

### `NaN`/`nan` 

Only `NaN`/`nan` is native in Python, if you cast it as a `float` first. 

```python
print(type(float('NaN')))
print(type(float('nan')))
```

But you can also import from the `math` package (which should already be installed in Python)

(`numpy` also has `nan`)

```python
import math

print(type(math.nan))
```

Testing logicals with `NaN`/`nan`

`math.nan == math.nan` is False because `nan` is not equal to anything, including itself. 

```python
math.nan == math.nan
```

Usually you test if an object is `nan` with:

```python
x = math.nan
math.isnan(x)
```

### `NA`

`NA` is from the `pandas` package, an important package for dealing with tabular data. 

```python
import pandas as pd

print(type(pd.NA))
```

Test status with:

```python
print(pd.isna(pd.NA))
print(type(pd.isna(pd.NA)))
```
